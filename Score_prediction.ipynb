# **Runs scored till the current ball**

```python
cdf['total_score']=cdf.groupby(['id', 'inning'])['total_runs'].apply(lambda x: x.cumsum())
```

```python
cdf['total_score'].head(20)
```

# **Runs scored in previous 30 balls**

```python
tmp=cdf.groupby(['id','inning'])['total_runs'].rolling(min_periods=1, window=30).sum().reset_index()
```

```python
cdf['prev_30_runs']=tmp['total_runs'].to_list()
```

```python
cdf['prev_30_runs'].head(31)
```

# **Wickets fallen in previous 30 balls**

```python
cdf['player_dismissed']=np.where(cdf['player_dismissed']==0, 0, 1)
```

```python
tmp = cdf.groupby(['id', 'inning'])['player_dismissed'].rolling(min_periods = 1, window = 30).sum().reset_index()
```

```python
cdf['prev_30_wickets'] = tmp['player_dismissed'].to_list()
```

```python
cdf.head(40)
```

```python
 cdf['total_wickets'] = cdf.groupby(['id', 'inning'])['player_dismissed'].apply(lambda x : x.cumsum())
```

# **Dot balls in previous 30 balls**

```python
cdf['prev_30_dot_balls']=cdf['total_runs']
cdf['prev_30_dot_balls']=np.where(cdf['prev_30_dot_balls']==0, 1, 0)

tmp=cdf.groupby(['id','inning'])['prev_30_dot_balls'].rolling(min_periods=1, window=30).sum().reset_index()
cdf['prev_30_dot_balls']=tmp['prev_30_dot_balls'].to_list()
```

# **Boundaries in previous 30 balls**


```python
cdf['prev_30_boundaries']=cdf['total_runs']
cdf['prev_30_boundaries']=np.where(cdf['prev_30_boundaries']>3, 1, 0)

tmp=cdf.groupby(['id','inning'])['prev_30_boundaries'].rolling(min_periods=1, window=30).sum().reset_index()
cdf['prev_30_boundaries']=tmp['prev_30_boundaries'].to_list()
```

# **train_test_split**

```python
X = temp.drop(labels=['total','id'], axis=1)
y = temp['total'].values
```

```python
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42,stratify=y)
```

```python
X_train=X_train.values
X_test=X_test.values
X_train=np.asarray(X_train).astype(np.float32)
X_test=np.asarray(X_test).astype(np.float32)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# **Training RF model**

```python
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 42)
  
# fit the regressor with x and y data
regressor.fit(X_train, y_train)  
```

```python
filename = '/content/drive/MyDrive/Cricket_Analytics/rf_cricket-model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```

```python
prediction = regressor.predict(X_test)
```

```python
print(mean_absolute_error(y_test, prediction))
```
